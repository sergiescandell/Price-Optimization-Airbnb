# Contexto

### El creciente número de apartamentos turísticos en Madrid hace que sea un mercado cada vez más competitivo. Con el aumento de la oferta, es fundamental identificar las características que tienen aquellos apartamentos con mayor ocupación y precio.
````python
# Leo el documento de propiedades listadas
df_listados = pd.read_csv("reviews.csv")
df_listados
````

````python
# Trato la fecha
df_listados["date"] = pd.to_datetime(df_listados["date"], format='%Y-%m-%d')

# Creo una columna con el año de la fecha
df_listados["year"] = df_listados["date"].dt.year
````

````python
listings_per_year = df_listados.groupby("year")["listing_id"].nunique().reset_index(name="num_listings")
listings_per_year
````

````python
ax = sns.barplot(listings_per_year, 
                 x = "year", 
                 y = "num_listings", 
                 hue = "year",
                 palette= "crest")

plt.legend().remove()

# Rotamos los años para que se vean mejor
plt.xticks(rotation = 45)

# Nombres variables y título del gráfico
plt.xlabel("Año")
plt.ylabel("Nº de nuevos apartamentos listados en Airbnb")
plt.title("Evolución nuevos apartamentos listados en Airbnb", fontsize=12, fontweight="bold")

# Escribir el numero de nuevos apartamentos listados cada año
for index, row in listings_per_year.iterrows():
    ax.text(index, row["num_listings"], f"{row['num_listings']}", 
            ha="center", va="bottom", fontsize=10)

# Personalizar el marco del plot 
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_color("gray")

plt.show()

print("En este gráfico se puede observar la evolución de nuevos apartamentos listados en Airbnb")
````

# Problema

### ¿Qué características o variables tienen un mayor impacto en las diferencias de precio por noche en apartamentos que tienen entre 1 y 3 habitaciones y entre 1 y 3 baños en la ciudad de Madrid, con un precio máximo de 200€?

# Información

````python
df_host = pd.read_parquet(r"Data/airbnb_madrid_host.parquet")

df_host.head(10).transpose()
````

````python
df_host["host_response_time"].isna().sum()
````

````python
# Rellenar los nulos de la columna host_response_time 
df_host.host_response_time = df_host.host_response_time.fillna("Unknown")
````

````python
df_location = pd.read_parquet(r"Data/airbnb_madrid_location.parquet")
df_location.head(10).transpose()
````

````python
df_location.neighbourhood.value_counts()
````

````python
df_property = pd.read_parquet("Data/airbnb_madrid_property.parquet")
df_property.head(10).transpose()
````

````python
df_property.bedrooms.value_counts()
````

````python
df_property.bathrooms.value_counts() 
````

Hay un apartamento con diez baños

````python
df_property.bed_type.value_counts()
````

````python
df_conditions = pd.read_parquet("Data/airbnb_madrid_conditions.parquet")
df_conditions.head(10).transpose()
````

````python
# Como queremos hacer un análisis de menos de 200 € por noche , hacemos un filtro al df
df_conditions = df_conditions[df_conditions["price"]<= 200]
````

````python
df_reviews = pd.read_parquet(r"Data/airbnb_madrid_reviews.parquet")
df_reviews.head(10).transpose()
````

````python
# Filtro apartamentos para que me queden solo los que tienen las mismas habitaciones y baños que los del propietario
filtro_habitaciones = (df_property.bedrooms >= 1) & (df_property.bedrooms <= 3) 
df_property_filtro_habitaciones = df_property[filtro_habitaciones]

filtro_baños = (df_property_filtro_habitaciones.bathrooms >=1) & (df_property_filtro_habitaciones.bathrooms <=3)
df_property_filtro_habitaciones_baños = df_property_filtro_habitaciones[filtro_baños]
````

````python
# Crep merge entre conditions y property filtrado para poder comparar.
df_property_conditions = pd.merge(left = df_property_filtro_habitaciones_baños, right = df_conditions, on = "id", how = "inner")
````

````python
df_property_conditions_host = pd.merge( left = df_property_conditions, right = df_host, on = "id", how = "inner")
````

````python
df_all = pd.merge(left = df_property_conditions_host, right = df_location, on = "id", how = "inner")
````

````python
df_all = pd.merge(left = df_all, right = df_reviews, on = "id", how = "inner")
````

### ¿Influye el tiempo de respuesta al precio?

````python
orden = ["within_an_hour", "within_a_few_hours", "within_a_day", "a_few_days_or_more", "Unknown"]
sns.boxplot(data = df_all,
            x = "host_response_time",
            y = "price",
            hue = "host_response_time",
            order = orden, 
            palette= "crest")

plt.xticks(rotation = 90)

plt.show()
````

````python
plt.figure(figsize=(4,6))
g = sns.catplot(data = df_all,
            x = "host_response_time",
            y = "price",
            kind = "bar",
            order = orden,
            hue = "host_is_superhost", 
            errorbar = None,
            palette= {0 : "#D84040", 1: "#2E5077"}
            )

# Personalizamos la leyenda de Seaborn 
g._legend.set_title("Host is Superhost") 
g._legend.set_frame_on(False)  
g._legend.get_texts()[0].set_text("False")  
g._legend.get_texts()[1].set_text("True") 

#Rotamos el nombre de las variables
plt.xticks(rotation = 90)

plt.xlabel("Tiempo de Respuesta del Host")
plt.ylabel("Precio")
plt.title("¿Cómo afecta el tiempo de respuesta al precio?", fontsize = 12, fontweight= "bold")

plt.show()
````

- `Insight` : El tiempo de respuesta del host no afecta al precio por noche
- `Insight` : Si tardan en responder lo mismo, si son superhost el precio por noche es más alto.

## ¿Qué barrio tiene más proporción de apartamentos dentro del 20% con mayor precio?

````python
valor_q80 = df_all.price.quantile(0.8)
print(f"cuantil 80 del precio de un apartamento es: {valor_q80}")
df_all["is_top20"] = df_all.price.gt(valor_q80)
````

````python
# seleccionamos los barrios para los que tenemos al menos 50 apartamentos
top_neighbourhood = (
    df_all
    .value_counts("neighbourhood_group")
    .sort_values(ascending=False)
    .to_frame(name="n")
    .query("n > 50")
    .index
)
# filtramos barrios y contamos la proporción de apartamentos top vs no-top
df_ = (
    df_all
    .query("neighbourhood.isin(@top_neighbourhood)")
    .groupby("neighbourhood", as_index=False)
    .value_counts(["is_top20"], normalize=True)
    .sort_values("proportion")
)
# barplot
px.bar(
    data_frame=df_,
    x="proportion",
    y="neighbourhood",
    color="is_top20",
    barmode="relative")
````

``Insight``: el barrio que más apartamentos con mayor precio tiene es Salamanca
